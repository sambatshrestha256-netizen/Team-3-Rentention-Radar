```python
X = df[['tenure', 'MonthlyCharges']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42
    )
    
    kmeans.fit(X_scaled)
    
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Customer Segmentation")

plt.show()
```

```python
import pandas as pd

df = pd.read_csv("stage1_cleaned_dataset.csv")

df.head()
```

```python
X = df[['tenure', 'MonthlyCharges']]

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42
    )
    
    kmeans.fit(X_scaled)
    
    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Customer Segmentation")

plt.show()
```

```python
from sklearn.cluster import KMeans

# Create K-Means model
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

# Train model and assign clusters
clusters = kmeans.fit_predict(X_scaled)

# Check first 10 cluster labels
clusters[:10]
```

```python
df['Cluster'] = clusters

df.head()
```

```python
cluster_summary = df.groupby('Cluster')[['tenure','MonthlyCharges']].mean()

cluster_summary
```

```python
df['Cluster'].value_counts()
```

```python
cluster_summary
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5))

plt.scatter(
    df['tenure'],
    df['MonthlyCharges'],
    c=df['Cluster']
)

plt.xlabel("Tenure")
plt.ylabel("Monthly Charges")
plt.title("Customer Segmentation using K-Means")

plt.show()
```
